```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./DATA/IMDb Movies India.csv", encoding="ISO-8859-1")
df.head()
```

```python
print(df.info())

```

```python
print(df.describe())
```

```python
sns.histplot(df['Rating'], bins=10, kde=True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
top_directors = df['Director'].value_counts().head(10).index
filtered_data = df[df['Director'].isin(top_directors)]

plt.figure(figsize=(12, 6))
sns.barplot(x='Director', y='Rating', data=filtered_data, errorbar=None)
plt.title('Average Ratings by Top Directors')
plt.xlabel('Director')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='Votes', y='Rating', data=df)
plt.title('Votes vs. Rating')
plt.xlabel('Votes Revenue')
plt.ylabel('Rating')
plt.show()
```

```python
yearly_ratings = df.groupby('Year')['Rating'].mean().reset_index()

sns.lineplot(x='Year', y='Rating', data=yearly_ratings)
plt.title('Average Ratings Over Time')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```

```python
top_actors = df['Actor 1'].value_counts().head(10).index
filtered_data = df[df['Actor 1'].isin(top_actors)]

plt.figure(figsize=(12, 6))
sns.barplot(x='Actor 1', y='Rating', data=filtered_data)
plt.title('Average Ratings by Top Actor 1')
plt.xlabel('Actor 1')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```
